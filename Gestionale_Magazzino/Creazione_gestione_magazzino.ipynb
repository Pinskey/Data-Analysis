Cella con la classe articolo, che permette di:
- definire i vari attributi dell' oggetto articolo
- convertire i vari elementi in un dizionario

```python
class Articolo:
    """ Classe che rappresenta un articolo di magazzino. """
    
    def __init__(self,articolo,brand,quantità,colore,taglia,costo,materiale,note):
        """ 
        Attributi:
            articolo: rappresenta il nome dell'articolo
            brand: rappresenta il brand dell' articolo
            quantità: rappresenta la quantità presente in magazzino
            colore: rappresenta il colore dell' articolo
            taglia: rappresenta la taglia dell'articolo
            costo: rappresenta il costo unitario dell'articolo
            materiale: rappresenta il materiale di cui è fatto il prodotto
            note: racchiude eventuali informazioni aggiuntive
        """
        self.articolo = articolo
        self.brand = brand
        self.quantità = quantità
        self.colore = colore
        self.taglia = taglia
        self.costo = costo
        self.materiale = materiale
        self.note = note
    
    def dizionario(self):
        """ Funzione che converte un articolo della classe in un elemento di un dizionario. """
        diz_visualizzazione = {}
        diz_visualizzazione['articolo'] = self.articolo
        diz_visualizzazione['brand'] = self.brand
        diz_visualizzazione['quantità'] = self.quantità
        diz_visualizzazione['colore'] = self.colore
        diz_visualizzazione['taglia'] = self.taglia
        diz_visualizzazione['costo'] = self.costo
        diz_visualizzazione['materiale'] = self.materiale
        diz_visualizzazione['note'] = self.note
        
        return diz_visualizzazione  
    
```

Cella contenente la funzione filtro avanzato

Tale funzione prende in input tutti i criteri di ricerca inseriti e cerca i prodotti del magazzino che soddisfano tutti quanti tali criteri contemporaneamente

```python
def filtro_avanzato(lista_trovati, lista_da_trovare):
    """ Funzione che svolge il ruolo di filtro avanzato, andando a fare una ricerca di tutti quei 
    prodotti che soddisfano tutti i criteri di ricerca contemporaneamente"""
    # Lista dei risultati filtro avanzato
    lista_ricerca_avanzata = []
    # ciclo su tutti gli articoli
    for articolo in lista_trovati:
        # contatore per conferma presenza di tutti i criteri
        contatore = 0
        # ciclo sugli item dei vari articoli
        for item in list(articolo.items()):
            # comparazione degli item dei vari articoli con le tuple salvate con la funzione di ricerca
            for trovato in lista_da_trovare:
                if len(trovato) == 2:
                    if trovato == item:
                        contatore += 1
                        if contatore == len(lista_da_trovare):
                            lista_ricerca_avanzata.append(articolo)  
                # Nel caso confronto costo e quantità
                elif len(trovato) == 3:
                    if trovato[2] == 'maggiore' and item[0]==trovato[0]:
                        if float(item[1]) > trovato[1]:
                            contatore += 1
                            if contatore == len(lista_da_trovare):
                                lista_ricerca_avanzata.append(articolo)
                    if trovato[2] == 'minore' and item[0]==trovato[0]:
                        if float(item[1]) < trovato[1]:
                            contatore += 1
                            if contatore == len(lista_da_trovare):
                                lista_ricerca_avanzata.append(articolo)  
                    if trovato[2] == 'uguale' and item[0]==trovato[0]:
                        if float(item[1]) == trovato[1]:
                            contatore += 1
                            if contatore == len(lista_da_trovare):
                                lista_ricerca_avanzata.append(articolo)  
    return lista_ricerca_avanzata
```

Di seguito le funzioni per poter estrarre delle statistiche utili ad una migliore comprensione del magazzino

```python
# Ho creato una serie di funzioni per avere qualche dato interessante
def media_prezzo(lista_magazzino):
    """ Funzione che calcola il prezzo medio dei prodotti del magazzino"""
    totale_costo = 0
    totale = 0
    for articolo in lista_magazzino:
        totale_costo += float(articolo.costo)*int(articolo.quantità)
        totale += int(articolo.quantità)
    return totale_costo/totale

def valore_totale_magazzino(lista_magazzino):
    """ Funzione che calcola il valore economico totale del magazzino"""
    valore_totale = 0
    for articolo in lista_magazzino:
        valore_totale += float(articolo.costo) * int(articolo.quantità)
    return valore_totale

def prezzo_max_min(lista_magazzino):
    """ Funzione che calcola prezzo massimo e minimo """
    lista_valori =[]
    for articolo in lista_magazzino:
        lista_valori.append(float(articolo.costo))
    return max(lista_valori),min(lista_valori)

def totale_articoli(lista_magazzino):
    """ Funzione che calcola il numero totale dei prodotti presenti in magazzino"""
    totale_articoli = 0
    for articolo in lista_magazzino:
        totale_articoli += int(articolo.quantità)
    return totale_articoli

def articolo_più_comune(lista_magazzino):
    """ Funzione che trova l'articolo più frequente """
    articolo_comune = []
    for articolo in lista_magazzino:
        articolo_comune.append(articolo.articolo)
    return max(set(articolo_comune),key = articolo_comune.count)

def brand_più_comune(lista_magazzino):
    """ Funzione che trova il brand più frequente """
    brand_comune = []
    for articolo in lista_magazzino:
        brand_comune.append(articolo.brand)
    return max(set(brand_comune),key = brand_comune.count)
```
